**statistics**

it means collecting data, organizing it and understanding it using numbers
it helps to find patterns or trend in data and make informed decisions


it is divided into 2 parts:


1. **descriptive stats** : descriptive stats means summarizing or describing the data in a way that is easier to understand

there are two types of descriptive stats:

1. **measure of central tedency** : it is a single value that attempts to describe the whole data set  

there are three main features of central tedency
1. mean
2. median
3. mode

2. **measure of despersion or varibility** : how spread out or scattered the values are in the dataset





2. **inferential stats** :   prediction and conclusion



```python
# 1.  measure of central tedency


# 1, mean
import statistics
student_marks  = [50,40,50,60,70,77,67,89,90,56]

print("the average marks of students is", statistics.mean(student_marks))
```

```python
# 2. median
data = [10,20,30,40,50,60,70]
print("median of a dataset is", statistics.median(data))
```

**2. measure of variability**

how much spread out or scatter the values are
how much the data values differ from each other


1. range
2. variance
3. standard deviation


```python
# 1 range :

# how spread out the data is
# it is difference between the largest and smallest number in the data

# formula = largest value  - smallest value
```

```python
marks = [40,30,50,60,55,46,76,67]

maximum = max(marks)
minimum  = min(marks)
range = maximum  - minimum
print("range of the data is ", range)
```
